```python
import pandas as pd

file_path = "squat_data.xlsx"

df = pd.read_excel(file_path)

print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

palette = {0: "blue", 1: "red"}

sns.scatterplot(x=df["UWB"], y=df["accel_x"], hue=df["updown"], palette=palette, ax=axes[0, 0]).set(title="UWB vs Accel X")
sns.scatterplot(x=df["UWB"], y=df["accel_y"], hue=df["updown"], palette=palette, ax=axes[0, 1]).set(title="UWB vs Accel Y")
sns.scatterplot(x=df["UWB"], y=df["accel_z"], hue=df["updown"], palette=palette, ax=axes[0, 2]).set(title="UWB vs Accel Z")
sns.scatterplot(x=df["UWB"], y=df["gyro_x"], hue=df["updown"], palette=palette, ax=axes[1, 0]).set(title="UWB vs Gyro X")
sns.scatterplot(x=df["UWB"], y=df["gyro_y"], hue=df["updown"], palette=palette, ax=axes[1, 1]).set(title="UWB vs Gyro Y")
sns.scatterplot(x=df["UWB"], y=df["gyro_z"], hue=df["updown"], palette=palette, ax=axes[1, 2]).set(title="UWB vs Gyro Z")

plt.tight_layout()
plt.show()
```

```python
df_down = df[df["updown"] == 0]
df_up = df[df["updown"] == 1]

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

palette = {0: "blue", 1: "red"}

sns.scatterplot(x=df_down["UWB"], y=df_down["accel_x"], hue=df_down["updown"], palette=palette, ax=axes[0, 0]).set(title="UWB vs Accel X")
sns.scatterplot(x=df_down["UWB"], y=df_down["accel_y"], hue=df_down["updown"], palette=palette, ax=axes[0, 1]).set(title="UWB vs Accel Y")
sns.scatterplot(x=df_down["UWB"], y=df_down["accel_z"], hue=df_down["updown"], palette=palette, ax=axes[0, 2]).set(title="UWB vs Accel Z")
sns.scatterplot(x=df_down["UWB"], y=df_down["gyro_x"], hue=df_down["updown"], palette=palette, ax=axes[1, 0]).set(title="UWB vs Gyro X")
sns.scatterplot(x=df_down["UWB"], y=df_down["gyro_y"], hue=df_down["updown"], palette=palette, ax=axes[1, 1]).set(title="UWB vs Gyro Y")
sns.scatterplot(x=df_down["UWB"], y=df_down["gyro_z"], hue=df_down["updown"], palette=palette, ax=axes[1, 2]).set(title="UWB vs Gyro Z")

plt.tight_layout()
plt.show()
```

```python
# -50 < accel_x < 50
# -50 < accel_y < 50
# -50 < accel_z < 50
# -50 < gyro_x < 50
# -50 < gyro_y < 50
# -50 < gyro_z < 50
# UWB < 2.2338
df_filtered = df[(df["accel_x"] > -30) & (df["accel_x"] < 30) & 
                 (df["accel_y"] > -30) & (df["accel_y"] < 30) & 
                 (df["accel_z"] > -30) & (df["accel_z"] < 30) & 
                 (df["gyro_x"] > -30) & (df["gyro_x"] < 30) & 
                 (df["gyro_y"] > -30) & (df["gyro_y"] < 30) & 
                 (df["gyro_z"] > -30) & (df["gyro_z"] < 30) & (df["UWB"] < 2.2338)]

df_filtered.describe()
```
