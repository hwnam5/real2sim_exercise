```python
import csv
import os
import sys
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('real_squat_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2,3, figsize=(15, 10))

sns.boxplot(x='updown', y='accel_x', data=df, ax=axes[0,0]).set_title('accel_x')
sns.boxplot(x='updown', y='accel_y', data=df, ax=axes[0,1]).set_title('accel_y')
sns.boxplot(x='updown', y='accel_z', data=df, ax=axes[0,2]).set_title('accel_z')
sns.boxplot(x='updown', y='gyro_x', data=df, ax=axes[1,0]).set_title('gyro_x')
sns.boxplot(x='updown', y='gyro_y', data=df, ax=axes[1,1]).set_title('gyro_y')
sns.boxplot(x='updown', y='gyro_z', data=df, ax=axes[1,2]).set_title('gyro_z')

plt.tight_layout()
plt.show()
```

```python
df_0_95 = df.iloc[:95]

colums = ['UWB','accel_x', 'accel_y', 'accel_z', 'gyro_x', 'gyro_y', 'gyro_z']

fig, axes = plt.subplots(2,4, figsize=(20, 15))

for i, col in enumerate(colums):
    ax = axes[i//4, i%4]
    ax.plot(df_0_95.index, df_0_95[col], marker='o', linestyle='-', color='blue')
    ax.set_title(col)

plt.tight_layout()
plt.show()
```

```python
df_0_95['accel_x'] = df_0_95['accel_x'].rolling(window=5).mean()
df_0_95['accel_y'] = df_0_95['accel_y'].rolling(window=5).mean()
df_0_95['accel_z'] = df_0_95['accel_z'].rolling(window=5).mean()
df_0_95['gyro_x'] = df_0_95['gyro_x'].rolling(window=5).mean()
df_0_95['gyro_y'] = df_0_95['gyro_y'].rolling(window=5).mean()
df_0_95['gyro_z'] = df_0_95['gyro_z'].rolling(window=5).mean()

fig, axes = plt.subplots(2,4, figsize=(20, 15))

for i, col in enumerate(colums):
    ax = axes[i//4, i%4]
    ax.plot(df_0_95.index, df_0_95[col], marker='o', linestyle='-', color='blue')
    ax.set_title(col)

plt.tight_layout()
plt.show()
```
